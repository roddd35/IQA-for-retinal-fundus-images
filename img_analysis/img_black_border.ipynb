```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
def make_image_square_with_black_borders(image_path):
    # load image
    original_image = Image.open(image_path)
    width, height = original_image.size
    print(f"Dimensões originais da imagem: {original_image.size}")
    
    # defines the new size (square with the biggest dimension)
    new_size = max(width, height)
    
    # create a new image with the dimensions wanted
    bordered_image = Image.new("RGB", (new_size, new_size), "black")
    
    # calculates the position to put the new image on the center
    left = (new_size - width) // 2
    top = (new_size - height) // 2
    
    # paste the image loaded onto the new black image
    bordered_image.paste(original_image, (left, top))

    # save the bordered_image
    # bordered_image.save('caminho/para/salvar/a/nova/imagem.jpg')
    
    return original_image, bordered_image

image_path = '/Users/rodrigomichelassi/Documents/USP/IQA-Motorola/algorithms/img_analysis/img00001.jpg'
original_image, bordered_image = make_image_square_with_black_borders(image_path)
# bordered_image.show()

print(f"Novas dimensões da imagem: {bordered_image.size}")

# plt.imshow(bordered_image)

f, (ax1, ax2) = plt.subplots(1, 2) 

ax1.set_title("Original Image")
ax1.set_xlabel("Horizontal size: %d" % original_image.size[0])
ax1.set_ylabel("Vertical size: %d" % original_image.size[1])
ax1.imshow(np.rot90(original_image, 2))

ax2.set_title("Bordered image")
ax2.set_xlabel("Horizontal size: %d" % bordered_image.size[0])
ax2.set_ylabel("Vertical size: %d" % bordered_image.size[1])
ax2.imshow(np.rot90(bordered_image, 2))

plt.subplots_adjust(wspace=0.5)
f.tight_layout()

plt.show()
```
