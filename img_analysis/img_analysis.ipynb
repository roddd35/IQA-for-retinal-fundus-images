```python
import os
from PIL import Image
import matplotlib.pyplot as plt
```

```python
path = '/Users/rodrigomichelassi/Documents/USP/IQA-Motorola/data/brset/physionet.org/files/brazilian-ophthalmological/1.0.0/fundus_photos'
```

```python
# criar um vetor com o nome de todas as imagens?
# pegar o nome do arquivo que abrimos, fazer as alteracoes e usar o mesmo nome para salvar em outro folder?
size_img = []

for archive in os.listdir(path):
    full_path = os.path.join(path, archive)
    if full_path.lower().endswith(('.png', '.jpg', '.jpeg')):
        with Image.open(full_path) as image:
            size_img.append(image.size)
            # size_img.append(img.copy()) crasha a aplicação
```

```python
print(size_img)
```

```python
list_sizes = []
list_occurencies = []
for x in size_img:
    if x not in list_sizes:
        list_sizes.append(x)
        list_occurencies.append(1)
    else:
        list_occurencies[list_sizes.index(x)] += 1
    
```

```python
print(list_sizes)
print(list_occurencies)
```
